```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/Resources_schools_complete.csv"

student_data_to_load = "Resources/Resources_students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
completed_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of students

* Calculate the average reading score

```python
# Calculate the total number of schools for both Schools and Student
school_count = len(completed_data["school_name"].unique())
school_count
```

```python
# Calculate the total number of schools for both Schools and Student
student_count = len(completed_data["Student ID"].unique())
student_count
```

```python
# Calculate the Total Budget
budget_total = school_data["budget"].sum()
budget_total
```

```python
# Calculate the average math score
math_avg = completed_data["math_score"].mean()
math_avg
```

```python
# Calculate the average reading score
reading_avg = completed_data["reading_score"].mean()
reading_avg
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)

# First get the count of students passing math
passing_math_stu_count = completed_data[(completed_data["math_score"] >= 70)].count()["student_name"]

passing_math_stu_count
```

```python
# Then calculate the percentage of students passing math & show the average
passing_math_stu_percent = passing_math_stu_count / float(student_count) * 100

passing_math_stu_percent
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)

# First get the count of students passing reading
passing_reading_stu_count = completed_data[(completed_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_stu_count
```

```python
# Then calculate the percentage of students passing reading & show the average
passing_reading_stu_percent = passing_reading_stu_count / float(student_count) * 100
passing_reading_stu_percent
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
count_stu_passed_both = completed_data[(completed_data["math_score"] >= 70) 
                                                  & (completed_data["reading_score"] >= 70)].count()["student_name"]

percent_overall_pass_rate = count_stu_passed_both /  float(student_count) * 100

percent_overall_pass_rate
```

```python
# Create a dataframe to hold the above results
# New dataframe will contain:
        #school_count
        #student_count
        #budget_total
        #math_avg
        #reading_avg
        #passing_math_stu_count
        #passing_math_stu_percent
        #passing_reading_stu_count
        #passing_reading_stu_percent
        #count_stu_passed_both
        #percent_overall_pass_rate

# Rename the columns of the new dataframe using the above variables created 

district_summary = pd.DataFrame({
    "Total Number of Schools": [school_count],
    "Total Number of Students": [student_count],
    "Budget Total": [budget_total],
    "Math Score Average": [math_avg],
    "Reading Score Average": [reading_avg],
    "% that Passed Math": [passing_math_stu_percent],
    "% that Passed Reading": [passing_reading_stu_percent],
    "% Rate that Passed Both Overall": [percent_overall_pass_rate]})

# Show the new columms and data
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create an index for School Name and type of school

school_types = school_data.set_index(["school_name"])["type"]

school_types
```

```python
# Determine how many students go to each school by pulling from completed_data using count
student_enrollment = completed_data["school_name"].value_counts()

student_enrollment
```

```python
# Calculate each school's Total School Budget; pulling from completed_data
# Must use a 'groupby' for school name and 'mean'for budget

budget_per_school = completed_data.groupby(["school_name"]).mean()["budget"]

budget_per_school
```

```python
# Calculate each school's Per Student Budget; pulling from completed_data
# Need to divide the "budget_per_school" by the "student_enrollment"
budget_per_student = budget_per_school / student_enrollment

budget_per_student
```

```python
# Calculate the Average Math Score; once again use 'groupby' for school name and 'mean'for math_score

avg_math_score_per_school = completed_data.groupby(["school_name"]).mean()["math_score"]


avg_math_score_per_school
```

```python
# Calculate the Average Reading Score; once again use 'groupby' for school name and 'mean'for reading_score
avg_reading_score_per_school = completed_data.groupby(["school_name"]).mean()["reading_score"]

avg_reading_score_per_school
```

```python
# Need to create new dataframes to separate out the students that passed math and reading.
# Easier to create two separate dataframes; one for math and another for reading

# new math df
stu_passed_math = completed_data[(completed_data["math_score"] >= 70)]

stu_passed_math
```

```python
# new math df

stu_passed_reading = completed_data[(completed_data["reading_score"] >= 70)]

stu_passed_reading
```

```python
# Using the new dataframes, calculate the % Passing Math
# divide the stu_passed_math by the student_enrollment; groupby on school_name column

pass_percent_school_math = stu_passed_math.groupby(["school_name"]).count()["student_name"] / student_enrollment * 100

pass_percent_school_math
```

```python
# Using the new dataframes, calculate the % Passing Reading
# divide the stu_passed_reading by the student_enrollment; groupby on school_name column

pass_percent_school_reading = stu_passed_reading.groupby(["school_name"]).count()["student_name"] / student_enrollment * 100

pass_percent_school_reading
```

```python
# Calculate the % Overall Passing 
#The percentage of students that passed math and reading.


# First make a new df that combines both math and reading. 
passing_both_subjects = completed_data[(completed_data["reading_score"] >= 70)
            & (completed_data["math_score"] >= 70)]

passing_both_subjects
```

```python
# Next calculate the % Overall Passing; use passing_both_subjects df
schools_pass_rate_both = passing_both_subjects.groupby(["school_name"]).count()["student_name"] / student_enrollment * 100

schools_pass_rate_both
```

```python
# Create a new dataframe to hold the above results
# New dataframe will contain:
        # school_types
        # student_enrollment
        # budget_per_school
        # budget_per_student
        # avg_math_score_per_school
        # avg_reading_score_per_school
        # stu_passed_math
        # stu_passed_reading
        # pass_percent_school_math
        # pass_percent_school_reading
        # passing_both_subjects
        # schools_pass_rate_both

# Rename the columns of the new dataframe using the above variables created 
# Don's use brackets like with the district summary or the table won't display correctly

summary_each_school = pd.DataFrame({
        "School Type": school_types, 
        "Student Enrollment": student_enrollment, 
        "Budget Per School": budget_per_school,
        "Budget Per Student": budget_per_student,
        "Average Math Score": avg_math_score_per_school,
        "Average Reading Score": avg_reading_score_per_school,
        "% Rate Passed Math": pass_percent_school_math,
        "% Rate Passed Reading": pass_percent_school_reading,
        "% Rate Passed Both": schools_pass_rate_both})
    

# Show the new columms
summary_each_school
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Use bins to create a new df and then categorize each using new bins
        # Use the budget per student from above
        # Bailey High School       628.0
        # Cabrera High School      582.0
        # Figueroa High School     639.0
        # Ford High School         644.0
        # Griffin High School      625.0
        # Hernandez High School    652.0
        # Holden High School       581.0
        # Huang High School        655.0
        # Johnson High School      650.0
        # Pena High School         609.0
        # Rodriguez High School    637.0
        # Shelton High School      600.0
        # Thomas High School       638.0
        # Wilson High School       578.0
        # Wright High School       583.0

budget_bins = [0, 585, 630, 650, 675]
group_names = ["<$584", "$585-629", "$630-649", "$650-675"]
```

```python
# Categorize spending based on the bins created.

summary_each_school["Spending Ranges (Per Student)"] = pd.cut(budget_per_school, budget_bins, labels=group_names)

summary_each_school
```

```python
#Average Math Score

spending_per_math_scores = summary_each_school.groupby([
    "Spending Ranges (Per Student)"]).mean()["Average Math Score"]
```

```python
#Average Reading Score

spending_per_reading_scores = summary_each_school.groupby([
    "Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
```

```python
#% Passing Math

spending_passed_math = summary_each_school.groupby([
    "Spending Ranges (Per Student)"]).mean()["% Rate Passed Math"]
```

```python
#% Passing Reading

spending_passed_reading = summary_each_school.groupby([
    "Spending Ranges (Per Student)"]).mean()["% Rate Passed Reading"]
```

```python
#Overall Passing Rate (Average of the above two)

overall_passed_spending = summary_each_school.groupby([
    "Spending Ranges (Per Student)"]).mean()["% Rate Passed Both"]
```

```python
# Make a new df with these spencing summaries

# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score" : spending_per_math_scores.round(2),
    "Average Reading Score": spending_per_reading_scores.round(2),
    "% Rate Passed Math": spending_passed_math.round(2),
    "% Rate Passed Reading": spending_passed_reading.round(2),
    "% Rate Passed Both": overall_passed_spending.round(2)})

spending_summary
```
